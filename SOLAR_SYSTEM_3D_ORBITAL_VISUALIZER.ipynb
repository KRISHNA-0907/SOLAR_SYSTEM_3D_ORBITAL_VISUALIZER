```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Define planets and their semi major axis and orbital period data
planet_names = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
semi_major_axis = [0.39, 0.72, 1.0, 1.52, 5.2, 9.58, 19.22, 30.05] #in AU
orbit_period = [0.24, 0.62, 1.0, 1.88, 11.86, 29.46, 84.01, 164.8] #in years

plt.style.use('dark_background')
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# plot sun as centre
ax.scatter(0, 0, 0, color='yellow', s=100, label='Sun')

# Defining orbit of each planets
for i, planet in enumerate(planet_names):
    orbit_radius = semi_major_axis[i]
    orbit_angle = np.linspace(0, 2*np.pi, 100)
    x = orbit_radius * np.cos(orbit_angle)
    y = orbit_radius * np.sin(orbit_angle)
    z = np.zeros_like(x)
    ax.plot(x, y, z, label=planet)

# plotting the data
ax.set_xlabel('X Distance(AU)')
ax.set_ylabel('Y Distance (AU)')
ax.set_zlabel('Z Distance (AU)')
ax.set_title('Solar System 3D visual')
ax.legend()
plt.show()

```
